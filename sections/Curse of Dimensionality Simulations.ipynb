```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display
import pylab as pl
import time
```

```python
def generate_examples(num_examples, num_dimensions):
    return np.random.uniform(0, 1, size=(num_examples, num_dimensions))
```

```python
ds = range(2, 200, 10)
f, axarr = plt.subplots(10, 2, figsize=(20, 75))
axarr = axarr.reshape(-1)
for i, d in enumerate(ds):
    X = generate_examples(1000, d)
    distances = distance.pdist(X, 'euclidean')
    axarr[i].set_title("d = " + str(d))
    axarr[i].hist(distances)
    axarr[i].set_xlim([0, np.sqrt(d)])
```
